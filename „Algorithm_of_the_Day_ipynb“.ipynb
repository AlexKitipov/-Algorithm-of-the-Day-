````python
# Algorithm of the Day
# --- Install Dependencies (Moved to the very top) ---
import sys
!{sys.executable} -m pip install gTTS
!{sys.executable} -m pip install pydub
!{sys.executable} -m pip install feedparser
print("‚úÖ gTTS, pydub, and feedparser installed successfully.")

import ipywidgets as widgets
from IPython.display import display, HTML, clear_output, Audio
import os
import datetime
import google.generativeai as genai
import requests
import feedparser
import random
import time
from tenacity import retry, stop_after_attempt, wait_exponential, retry_if_exception_type
from PIL import Image, ImageDraw, ImageFont # Import Pillow for image generation
from pydub import AudioSegment
from pydub.playback import play # For playing audio segments
from gtts import gTTS # Import gTTS for text-to-speech
import csv # Added import for csv module

# --- Global Setup ---
GOOGLE_API_KEY = "api_key=GOOGLE_API_KEY"
genai.configure(api_key=GOOGLE_API_KEY)

try:
    gemini_analyst_model = genai.GenerativeModel('gemini-2.0-flash')
    print("‚úÖ Gemini Flash Pro –º–æ–¥–µ–ª—ä—Ç –µ —É—Å–ø–µ—à–Ω–æ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–∞–Ω!")
except Exception as e:
    print(f"‚ùå –ì—Ä–µ—à–∫–∞ –ø—Ä–∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–∞–Ω–µ –Ω–∞ Gemini Flash Pro –º–æ–¥–µ–ª–∞: {e}")
    gemini_analyst_model = None # Set to None if initialization fails

project_folder_path = '/content/drive/MyDrive/Algorithm of the Day'
chat_logs_folder = os.path.join(project_folder_path, 'chat_logs')
music_current_folder = os.path.join(project_folder_path, 'music', 'current')
episodes_folder = os.path.join(project_folder_path, 'episodes')
journal_folder = os.path.join(project_folder_path, 'journal')
assets_covers_folder = os.path.join(project_folder_path, 'assets', 'covers')

try:
    os.makedirs(project_folder_path, exist_ok=True)
    os.makedirs(chat_logs_folder, exist_ok=True)
    os.makedirs(music_current_folder, exist_ok=True)
    os.makedirs(episodes_folder, exist_ok=True)
    os.makedirs(journal_folder, exist_ok=True)
    os.makedirs(assets_covers_folder, exist_ok=True)
    print(f"‚úÖ Successfully created or ensured existence of project folders.")
except Exception as e:
    print(f"‚ùå Error creating project folders: {e}")

# Placeholder for song_list
song_list = [
    {
        'url': "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-1.mp3",
        'filename': "SoundHelix-Song-1.mp3",
        'genre': "ambient"
    },
    {
        'url': "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-2.mp3",
        'filename': "SoundHelix-Song-2.mp3",
        'genre': "jazz"
    },
    {
        'url': "https://www.soundhelix.com/examples/mp3/SoundHelix-Song-3.mp3",
        'filename': "SoundHelix-Song-3.mp3",
        'genre': "electronic"
    }
]

# List of RSS feed URLs for diverse tech news
RSS_FEED_URLS = [
    "https://techcrunch.com/feed/",
    "http://feeds.bbci.co.uk/news/technology/rss.xml", # BBC Technology News
    "https://www.reuters.com/news/archive/technology" # Reuters Technology News (might need adjustment for direct RSS)
]

# --- Helper Functions (Audio Download, Playback, TTS) ---
def download_audio(url, folder_path, filename=None):
    """Downloads an audio file from a given URL to a specified folder."""
    if not filename:
        filename = url.split('/')[-1] if url.split('/')[-1] else "downloaded_audio.mp3"

    file_path = os.path.join(folder_path, filename)
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"‚úÖ Successfully downloaded '{filename}' to '{folder_path}'")
        return file_path
    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error downloading file from {url}: {e}")
        return None
    except Exception as e:
        print(f"‚ùå An unexpected error occurred: {e}")
        return None

def play_audio(file_path):
    """Plays an audio file from a given local path."""
    if os.path.exists(file_path):
        print(f"üé∂ Now playing: {os.path.basename(file_path)}")
        display(Audio(file_path, autoplay=True))
    else:
        print(f"‚ùå Audio file not found at '{file_path}'")

def text_to_audio(text, filename, output_folder, lang='bg'): # Default to Bulgarian
    """Converts text to speech using gTTS and saves it as an MP3 file.

    Args:
        text (str): The text to convert to speech.
        filename (str): The desired filename (e.g., 'gemini_response.mp3').
        output_folder (str): The folder where the audio file will be saved.
        lang (str): The language for the TTS (e.g., 'en' for English, 'bg' for Bulgarian).

    Returns:
        str: The full path to the saved audio file if successful, None otherwise.
    """
    try:
        tts = gTTS(text=text, lang=lang, slow=False)
        file_path = os.path.join(output_folder, filename)
        os.makedirs(output_folder, exist_ok=True)
        tts.save(file_path)
        print(f"‚úÖ Successfully converted text to audio and saved as '{file_path}' using gTTS")
        return file_path
    except Exception as e:
        print(f"‚ùå Error converting text to audio with gTTS: {e}")
        return None


# --- Helper Functions (Episode Generation Logic) ---
def select_and_store_song(songs):
    """
    Simulates Gemini selecting a song, downloads it, and plays it.

    Args:
        songs (list): A list of dictionaries, where each dictionary represents a song
                      with 'url', 'filename', and 'genre' keys.

    Returns:
        dict: The details of the selected and downloaded song.
    """
    if not songs:
        print("Error: Song list is empty.")
        return None, None

    selected_song = random.choice(songs)
    print(f"üß† Gemini selected: {selected_song['filename']} (Genre: {selected_song['genre']})")

    downloaded_file_path = download_audio(
        selected_song['url'],
        music_current_folder,
        selected_song['filename']
    )

    if downloaded_file_path:
        return selected_song, downloaded_file_path # Return both song details and path
    else:
        print(f"‚ùå Failed to download {selected_song['filename']}.")
        return None, None

def fetch_real_tech_news(rss_feed_urls=RSS_FEED_URLS):
    """Fetches the latest technology news headline from a randomly selected RSS feed."""
    random.shuffle(rss_feed_urls) # Shuffle to try different feeds

    for rss_feed_url in rss_feed_urls:
        try:
            response = requests.get(rss_feed_url, timeout=10)
            response.raise_for_status()
            feed = feedparser.parse(response.content)

            if feed.entries:
                latest_entry = feed.entries[0]
                title = latest_entry.title
                print(f"‚úÖ Successfully fetched news from RSS: {title} (Source: {rss_feed_url})")
                return title
            else:
                print(f"‚ùå RSS feed from {rss_feed_url} has no entries.")
        except requests.exceptions.RequestException as e:
            print(f"‚ùå Network error fetching RSS feed from {rss_feed_url}: {e}")
        except Exception as e:
            print(f"‚ùå An unexpected error occurred while fetching/parsing RSS feed from {rss_feed_url}: {e}")

    print("‚ùå All attempts to fetch tech news from available RSS feeds failed.")
    return "No recent tech news found due to multiple errors."


@retry(
    stop=stop_after_attempt(5), # Keep 5 attempts
    wait=wait_exponential(multiplier=1, min=4, max=15), # Exponential backoff with max 15s
    retry=retry_if_exception_type((genai.types.BlockedPromptException, requests.exceptions.ReadTimeout)) # Specific exceptions
)
def _generate_content_with_retry(model, prompt):
    return model.generate_content(prompt, request_options={'timeout': 120})

def generate_commentary(news_event):
    """Generates a brief comment and a metaphor based on a news event using Gemini."""
    global gemini_analyst_model

    if gemini_analyst_model is None:
        print("‚ùå Error: 'gemini_analyst_model' is not initialized. Cannot generate commentary.")
        return None, None

    prompt = f"""
    Analyze the following tech news. Provide a *very concise* comment (max 2 sentences) and a *short, creative* metaphor.
    Output format:
    Comment: [Your brief comment]
    Metaphor: [Your creative metaphor]

    News: {news_event}
    """

    try:
        response = _generate_content_with_retry(gemini_analyst_model, prompt)
        gemini_response = response.text
        print("\n--- Gemini Raw Response ---")
        print(gemini_response)
        print("-------------------------")

        comment = None
        metaphor = None

        response_lines = gemini_response.split('\n')
        for line in response_lines:
            if line.startswith('Comment:'):
                comment = line.replace('Comment:', '').strip()
            elif line.startswith('Metaphor:'):
                metaphor = line.replace('Metaphor:', '').strip()

        return comment, metaphor

    except Exception as e:
        print(f"‚ùå Error communicating with Gemini or parsing response after multiple retries: {e}")
        return None, None

def normalize_audio(audio_segment, target_dbfs=-20.0):
    """Normalizes an AudioSegment to a target dBFS level."""
    change_in_db = target_dbfs - audio_segment.dBFS
    normalized_segment = audio_segment.apply_gain(change_in_db)
    return normalized_segment

def fade_audio(audio_segment, fade_in_ms=500, fade_out_ms=1000):
    """Applies fade-in and fade-out effects to an AudioSegment."""
    faded_segment = audio_segment.fade_in(fade_in_ms).fade_out(fade_out_ms)
    return faded_segment

def mix_audio_segments(background_audio, foreground_audio, start_position_ms, volume_reduction_db=6):
    """Mixes a foreground audio segment onto a background audio segment.

    Args:
        background_audio (AudioSegment): The background audio segment.
        foreground_audio (AudioSegment): The foreground audio segment.
        start_position_ms (int): The start time in milliseconds for the foreground audio.
        volume_reduction_db (int): How much to reduce the background volume during foreground playback.

    Returns:
        AudioSegment: The combined audio segment.
    """
    if len(background_audio) < start_position_ms + len(foreground_audio):
        background_audio += AudioSegment.silent(duration=start_position_ms + len(foreground_audio) - len(background_audio), frame_rate=background_audio.frame_rate)

    background_part_before = background_audio[:start_position_ms]
    background_part_during_foreground = background_audio[start_position_ms : start_position_ms + len(foreground_audio)] - volume_reduction_db
    background_part_after = background_audio[start_position_ms + len(foreground_audio):]

    processed_background = background_part_before + background_part_during_foreground + background_part_after
    combined_audio = processed_background.overlay(foreground_audio, position=start_position_ms)
    return combined_audio

def generate_episode_cover_image(news_event, metaphor, genre):
    """Generates a placeholder episode cover image file path and optionally a dummy image.

    Args:
        news_event (str): The news headline/summary.
        metaphor (str): The metaphor generated for the episode.
        genre (str): The genre of the chosen song.

    Returns:
        str: The full path to the generated cover image file (or placeholder).
    """
    print(f"üñºÔ∏è Generating cover for news: '{news_event}' (Genre: {genre})")

    current_datetime = datetime.datetime.now()
    timestamp = current_datetime.strftime("%Y%m%d_%H%M%S")
    cover_filename = f"cover_{timestamp}.png"
    cover_image_path = os.path.join(assets_covers_folder, cover_filename)

    try:
        img = Image.new('RGB', (1200, 675), color = (73, 109, 137))
        d = ImageDraw.Draw(img)
        try:
            fnt_title = ImageFont.truetype('arial.ttf', 50)
            fnt_body = ImageFont.truetype('arial.ttf', 35)
        except IOError:
            fnt_title = ImageFont.load_default()
            fnt_body = ImageFont.load_default()

        date_text = current_datetime.strftime("%d.%m.%Y")
        d.text((50, 50), f"–î–ê–¢–ê: {date_text}", font=fnt_body, fill=(255, 255, 255))

        display_news = news_event[:70] + '...' if len(news_event) > 70 else news_event
        display_metaphor = metaphor[:70] + '...' if len(metaphor) > 70 else metaphor

        text_lines = [
            "–ê–ª–≥–æ—Ä–∏—Ç—ä–º–∞ –Ω–∞ –¥–µ–Ω—è",
            f"–¢–µ–º–∞: {display_news}",
            f"–ú–µ—Ç–∞—Ñ–æ—Ä–∞: {display_metaphor}",
            f"–ú—É–∑–∏–∫–∞–ª–µ–Ω –ñ–∞–Ω—Ä: {genre.capitalize()}"
        ]
        y_text = 180
        for i, line in enumerate(text_lines):
            if i == 0:
                d.text((50, y_text), line, font=fnt_title, fill=(255, 255, 255))
                y_text += 70
            else:
                d.text((50, y_text), line, font=fnt_body, fill=(255, 255, 255))
                y_text += 60

        img.save(cover_image_path)
        print(f"‚úÖ Generated cover image at '{cover_image_path}'")
        return cover_image_path
    except ImportError:
        print("‚ùå PIL (Pillow) not installed. Cannot generate actual image. Returning placeholder path.")
        with open(cover_image_path, 'w') as f:
            f.write("Placeholder content for cover image")
        return cover_image_path
    except Exception as e:
        print(f"‚ùå Error generating cover image: {e}. Returning placeholder path.")
        with open(cover_image_path, 'w') as f:
            f.write("Placeholder content for cover image")
        return cover_image_path

def archive_episode(news_event, comment, metaphor, chosen_song, final_mixed_audio_path, cover_image_path):
    """Archives the complete daily episode content into structured files, including the final mixed audio."""
    try:
        timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        episode_filename_base = f"episode_{timestamp}"
        episode_markdown_filename = f"{episode_filename_base}.md"
        episode_markdown_path = os.path.join(episodes_folder, episode_markdown_filename)

        markdown_content = f"""# –ê–ª–≥–æ—Ä–∏—Ç—ä–º –Ω–∞ –¥–µ–Ω—è: {news_event}\n\n## –ü—Ä–æ–ª–æ–≥\n\n"""

        markdown_content += f"""## –§–∞–∫—Ç\n\n**–ù–æ–≤–∏–Ω–∞:** {news_event}\n\n## –ú–µ—Ç–∞—Ñ–æ—Ä–∞\n\n**–ú–µ—Ç–∞—Ñ–æ—Ä–∞:** {metaphor}\n\n## –ö–æ–º–µ–Ω—Ç–∞—Ä\n\n**–ö–æ–º–µ–Ω—Ç–∞—Ä –æ—Ç Gemini:** {comment}\n\n## –ú—É–∑–∏–∫–∞\n\n**–ò–∑–±—Ä–∞–Ω–∞ –ø–µ—Å–µ–Ω:** {chosen_song.get('filename', 'N/A')} (–ñ–∞–Ω—Ä: {chosen_song.get('genre', 'N/A')})\n**–õ–∏–Ω–∫:** {chosen_song.get('url', 'N/A')})\n\n## –ê—É–¥–∏–æ –Ω–∞ –ï–ø–∏–∑–æ–¥–∞\n\n**–ü—ä–ª–µ–Ω –∞—É–¥–∏–æ –µ–ø–∏–∑–æ–¥:** {os.path.basename(final_mixed_audio_path)}\n\n## –ö–æ—Ä–∏—Ü–∞\n\n**–ò–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –Ω–∞ –∫–æ—Ä–∏—Ü–∞—Ç–∞:** {os.path.basename(cover_image_path)}\n\n## –†–∏—Ç—É–∞–ª\n\n"""

        markdown_content += f"""## –í—ä–ø—Ä–æ—Å –∫—ä–º –æ–±—â–Ω–æ—Å—Ç—Ç–∞\n\n"""

        with open(episode_markdown_path, 'w', encoding='utf-8') as f:
            f.write(markdown_content)
        print(f"‚úÖ Episode Markdown saved to: '{episode_markdown_path}'")

        journal_csv_path = os.path.join(journal_folder, 'journal.csv')
        file_exists = os.path.isfile(journal_csv_path)

        episode_key = episode_filename_base

        with open(journal_csv_path, 'a', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['–∫–ª—é—á', '–¥–∞—Ç–∞', '–∑–∞–≥–ª–∞–≤–∏–µ', '–∫–æ–º–µ–Ω—Ç–∞—Ä', '–ø–µ—Å–µ–Ω', '–∫–æ—Ä–∏—Ü–∞', '–∞—É–¥–∏–æ_–µ–ø–∏–∑–æ–¥']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            if not file_exists:
                writer.writeheader()

            writer.writerow({
                '–∫–ª—é—á': episode_key,
                '–¥–∞—Ç–∞': datetime.datetime.now().isoformat(),
                '–∑–∞–≥–ª–∞–≤–∏–µ': news_event,
                '–∫–æ–º–µ–Ω—Ç–∞—Ä': comment,
                '–ø–µ—Å–µ–Ω': chosen_song.get('filename', 'N/A'),
                '–∫–æ—Ä–∏—Ü–∞': os.path.basename(cover_image_path),
                '–∞—É–¥–∏–æ_–µ–ø–∏–∑–æ–¥': os.path.basename(final_mixed_audio_path)
            })
        print(f"‚úÖ Episode details appended to: '{journal_csv_path}'")
        return True

    except Exception as e:
        print(f"‚ùå Error archiving episode: {e}")
        return False

# --- Main play_show function ---
def play_show():
    """Orchestrates the entire daily episode generation and presentation logic, including audio mixing."""
    print("\n--- Starting '–ê–ª–≥–æ—Ä–∏—Ç—ä–º–∞ –Ω–∞ –¥–µ–Ω—è' Episode Generation ---")

    # 1. Simulate Gemini 'asking' for news
    print("üß† Gemini: '–ö–∞–∫–≤–æ –Ω–æ–≤–æ –¥–Ω–µ—Å –≤ —Å–≤–µ—Ç–∞ –Ω–∞ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏–∏—Ç–µ?' (What's new today in the world of technology?)")

    # 2. Fetch the latest technology news event
    news_event = fetch_real_tech_news()
    if not news_event:
        print("‚ùå Could not fetch news, aborting episode generation.")
        return

    # 3. Generate news audio (Gemini reading) using gTTS
    print("üß† Gemini: '–ù–µ–∫–∞ –¥–∞ —á—É–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏—Ç–µ –Ω–æ–≤–∏–Ω–∏...' (Let's hear the latest news...)")
    news_audio_filename = "news_headline.mp3"
    news_audio_file_path = text_to_audio(news_event, news_audio_filename, music_current_folder, lang='bg') # Using Bulgarian for Gemini's voice
    if not news_audio_file_path:
        print("‚ùå Failed to generate audio for news event.")
        return

    # 4. Generate commentary
    print("\nü§ñ Copilot: '–†–∞–∑–±—Ä–∞—Ö, –µ—Ç–æ –º–æ—è—Ç –ø–æ–≥–ª–µ–¥ –∏ –º–µ—Ç–∞—Ñ–æ—Ä–∞...' (Understood, here's my perspective and metaphor...)")
    comment, metaphor = generate_commentary(news_event)

    # --- Fallback mechanism ---
    if not comment or not metaphor:
        print("‚ö†Ô∏è Gemini failed to generate commentary. Using Copilot fallback.")
        comment = "Copilot: –ù–æ–≤–∏–Ω–∞—Ç–∞ –ø–æ–∫–∞–∑–≤–∞ –∫–æ–ª–∫–æ —Ç—Ä—É–¥–Ω–æ –µ –¥–∞ —Å–µ –∏–∑–º–µ—Ä–∏ –Ω–µ–ø–æ–∑–Ω–∞—Ç–æ—Ç–æ. –¢–æ–≤–∞ –µ –∫–∞—Ç–æ –¥–∞ —Å–µ –æ–ø–∏—Ç–∞—à –¥–∞ –∑–∞—Å—Ç—Ä–∞—Ö–æ–≤–∞—à –±—É—Ä—è ‚Äì –∑–Ω–∞–µ—à, —á–µ –∏–¥–≤–∞, –Ω–æ –Ω–µ –∑–Ω–∞–µ—à —Å–∏–ª–∞—Ç–∞ –π."
        metaphor = "–¢–æ–≤–∞ –µ –∫–∞—Ç–æ –¥–∞ —Å–µ –æ–ø–∏—Ç–∞—à –¥–∞ –∑–∞—Å—Ç—Ä–∞—Ö–æ–≤–∞—à —á–µ—Ä–Ω–∞ –¥—É–ø–∫–∞ ‚Äì –∑–Ω–∞–µ—à, —á–µ –µ –º–æ—â–Ω–∞, –Ω–æ –Ω–µ –º–æ–∂–µ—à –¥–∞ –ø—Ä–µ–¥–≤–∏–¥–∏—à –∫–∞–∫–≤–æ —â–µ –ø–æ–≥—ä–ª–Ω–µ."

    # 5. Generate commentary audio using gTTS
    print(f"\nü§ñ Copilot –ö–æ–º–µ–Ω—Ç–∞—Ä: {comment}")
    print(f"ü§ñ Copilot –ú–µ—Ç–∞—Ñ–æ—Ä–∞: {metaphor}")

    comment_audio_filename = "comment_audio.mp3"
    comment_audio_file_path = text_to_audio(comment, comment_audio_filename, music_current_folder, lang='bg') # Using Bulgarian for Copilot's voice
    if not comment_audio_file_path:
        print("‚ùå Failed to generate audio for commentary.")
        return

    # 6. Select and download a song for the '–ø–∞—É–∑–∞' (music break)
    print("\nüé∂ Gemini: '–í—Ä–µ–º–µ –∑–∞ –º—É–∑–∏–∫–∞–ª–Ω–∞ –ø–∞—É–∑–∞.' (Time for a music break.)")
    chosen_song, downloaded_song_path = select_and_store_song(song_list)
    if not chosen_song or not downloaded_song_path:
        print("‚ùå Could not select and store song, aborting episode generation.")
        return

    # --- Audio Processing with Pydub ---
    print("\n--- Mixing Audio Segments ---")
    try:
        # Load all audio segments
        news_segment = AudioSegment.from_file(news_audio_file_path)
        comment_segment = AudioSegment.from_file(comment_audio_file_path)
        music_segment = AudioSegment.from_file(downloaded_song_path)

        # Ensure all segments have consistent frame rate and channels for mixing
        target_frame_rate = 44100
        target_channels = 2 # Stereo for music, converting others

        news_segment = news_segment.set_frame_rate(target_frame_rate).set_channels(target_channels)
        comment_segment = comment_segment.set_frame_rate(target_frame_rate).set_channels(target_channels)
        music_segment = music_segment.set_frame_rate(target_frame_rate).set_channels(target_channels)

        # Normalize audio segments
        news_segment = normalize_audio(news_segment, target_dbfs=-15.0)
        comment_segment = normalize_audio(comment_segment, target_dbfs=-15.0)
        music_segment = normalize_audio(music_segment, target_dbfs=-25.0) # Background music should be quieter

        # Apply fade to music
        music_segment = fade_audio(music_segment, fade_in_ms=3000, fade_out_ms=5000)

        # Start building the final episode audio
        final_episode_audio = AudioSegment.silent(duration=1000, frame_rate=target_frame_rate) # 1 sec intro silence
        final_episode_audio = final_episode_audio.set_channels(target_channels) # Set channels after creation

        # Add News Segment
        final_episode_audio += news_segment
        final_episode_audio += AudioSegment.silent(duration=500, frame_rate=target_frame_rate) # Short pause
        final_episode_audio = final_episode_audio.set_channels(target_channels) # Set channels after creation

        # Add Commentary Segment (overlayed on ducked music for a short while)
        music_intro_duration = 5000 # 5 seconds of music before comment
        comment_start_time_ms = len(final_episode_audio) # Start where news ended + pause

        music_under_comment = music_segment[comment_start_time_ms : comment_start_time_ms + len(comment_segment) + music_intro_duration]

        mixed_commentary = mix_audio_segments(
            music_under_comment,
            comment_segment,
            start_position_ms=music_intro_duration,
            volume_reduction_db=10
        )

        final_episode_audio += mixed_commentary
        final_episode_audio += AudioSegment.silent(duration=1000, frame_rate=target_frame_rate) # Pause after commentary
        final_episode_audio = final_episode_audio.set_channels(target_channels) # Set channels after creation

        # Add remaining music (fading out)
        remaining_music = music_segment[len(final_episode_audio):]
        final_episode_audio += remaining_music

        # Export final mixed audio
        timestamp_for_audio = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
        final_mixed_audio_filename = f"episode_full_mix_{timestamp_for_audio}.mp3"
        final_mixed_audio_path = os.path.join(music_current_folder, final_mixed_audio_filename)
        final_episode_audio.export(final_mixed_audio_path, format="mp3")

        print(f"‚úÖ Final mixed episode audio exported to: '{final_mixed_audio_path}'")
        play_audio(final_mixed_audio_path) # Play the final mixed audio

    except Exception as e:
        print(f"‚ùå Error during audio mixing: {e}")
        final_mixed_audio_path = None # Indicate failure to mix audio

    if not final_mixed_audio_path:
        print("‚ùå Failed to create final mixed audio, aborting archiving.")
        return

    # 7. Generate episode cover image
    print("\nüñºÔ∏è Generating episode cover image...")
    cover_image_path = generate_episode_cover_image(news_event, metaphor, chosen_song.get('genre', 'general'))
    if not cover_image_path:
        print("‚ùå Failed to generate cover image.")

    # 8. Archive all generated content, using the final mixed audio path
    print("\nüíæ Archiving today's episode...")
    archived_successfully = archive_episode(
        news_event=news_event,
        comment=comment,
        metaphor=metaphor,
        chosen_song=chosen_song,
        final_mixed_audio_path=final_mixed_audio_path,
        cover_image_path=cover_image_path
    )

    if archived_successfully:
        print("üéâ Episode archiving completed successfully.")
    else:
        print("‚ùå Episode archiving failed.")

    # Clean up individual audio files after mixing and archiving
    try:
        if os.path.exists(news_audio_file_path): os.remove(news_audio_file_path)
        if os.path.exists(comment_audio_file_path): os.remove(comment_audio_file_path)
        if os.path.exists(downloaded_song_path): os.remove(downloaded_song_path)
        print("‚úÖ Cleaned up temporary individual audio files.")
    except Exception as e:
        print(f"‚ùå Error cleaning up temporary audio files: {e}")

    print("--- '–ê–ª–≥–æ—Ä–∏—Ç—ä–º–∞ –Ω–∞ –¥–µ–Ω—è' Episode Generation Complete ---\n")


# --- Chat Interface Widgets ---
chat_input = widgets.Textarea(
    value='',
    placeholder='Type your message here, or use /save [filename] [content] or /read [filename] or call play_show()',
    description='Your Message:',
    disabled=False,
    layout=widgets.Layout(width='auto', height='80px')
)

chat_output = widgets.Output(layout=widgets.Layout(border='1px solid black', width='auto', height='300px'))

send_button = widgets.Button(
    description='Send to Gemini',
    disabled=False,
    button_style='info',
    tooltip='Click to send your message to Gemini',
    icon='send'
)

# Global variable to store file content to be sent to Gemini
file_content_for_gemini = ""

# Function to handle sending messages and displaying responses
def on_send_button_clicked(b):
    global file_content_for_gemini
    with chat_output:
        clear_output(wait=True)
        user_message = chat_input.value
        chat_input.value = '' # Clear input after sending

        if not user_message.strip():
            chat_output.append_stdout("Please enter a message.\n")
            return

        chat_output.append_stdout(f"üßë‚Äçüíª You: {user_message}\n")

        # --- Handle /save, /read, and function calls ---
        if user_message.startswith('/save '):
            parts = user_message.split(' ', 2)
            if len(parts) >= 3:
                filename = parts[1]
                content_to_save = parts[2]
                file_path = os.path.join(project_folder_path, filename)
                try:
                    with open(file_path, 'w', encoding='utf-8') as f:
                        f.write(content_to_save)
                    chat_output.append_stdout(f"‚úÖ Saved '{filename}' to '{project_folder_path}'.\n")
                except Exception as e:
                    chat_output.append_stdout(f"‚ùå Error saving file '{filename}': {e}\n")
            else:
                chat_output.append_stdout("‚ùó Usage: /save [filename] [content]\n")
            file_content_for_gemini = "" # Clear any pending file content
            return

        elif user_message.startswith('/read '):
            parts = user_message.split(' ', 1)
            if len(parts) >= 2:
                filename = parts[1]
                file_path = os.path.join(project_folder_path, filename)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        content_read = f.read()
                    chat_output.append_stdout(f"‚úÖ Content of '{filename}':\n```\n{content_read}\n```\n")
                    file_content_for_gemini = content_read # Store content to send to Gemini
                    chat_output.append_stdout("‚ÑπÔ∏è File content will be included in the next message to Gemini.\n")
                except FileNotFoundError:
                    chat_output.append_stdout(f"‚ùå File '{filename}' not found in '{project_folder_path}'.\n")
                    file_content_for_gemini = ""
                except Exception as e:
                    chat_output.append_stdout(f"‚ùå Error reading file '{filename}': {e}\n")
                    file_content_for_gemini = ""
            else:
                chat_output.append_stdout("‚ùó Usage: /read [filename]\n")
            return

        elif user_message.strip() == 'play_show()':
            try:
                play_show()
                chat_output.append_stdout("‚úÖ play_show() executed successfully.\n")
            except Exception as e:
                chat_output.append_stdout(f"‚ùå Error executing play_show(): {e}\n")
            return

        # --- Construct prompt for Gemini, including file content if available ---
        gemini_prompt = user_message
        if file_content_for_gemini:
            gemini_prompt = f"Here is some file content for context:\n```\n{file_content_for_gemini}\n```\n\nMy request: {user_message}"
            file_content_for_gemini = "" # Clear after use

        chat_output.append_stdout("üß† Gemini is thinking...\n")

        try:
            if gemini_analyst_model is None:
                chat_output.append_stdout("‚ùå Error: 'gemini_analyst_model' is not initialized. Please ensure the setup cells ran correctly.\n")
                return

            response = gemini_analyst_model.generate_content(gemini_prompt)
            gemini_response = response.text
            chat_output.append_stdout(f"üß† Gemini: {gemini_response}\n")

            # --- Save Gemini's response to file ---
            timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
            text_filename = f"gemini_response_{timestamp}.txt"
            text_file_path = os.path.join(chat_logs_folder, text_filename)
            with open(text_file_path, 'w', encoding='utf-8') as f:
                f.write(gemini_response)
            chat_output.append_stdout(f"‚úÖ Gemini's response saved to '{text_file_path}'.\n")

            # --- Play Gemini's response aloud (inlined logic) ---
            chat_output.append_stdout("üß† Gemini: '–©–µ –ø—Ä–æ—á–µ—Ç–∞ –æ—Ç–≥–æ–≤–æ—Ä–∞ —Å–∏ –Ω–∞ –≥–ª–∞—Å.' (I will read my response aloud.)\n")
            audio_filename = f"gemini_response_audio_{timestamp}.mp3"
            audio_file_path = text_to_audio(gemini_response, audio_filename, chat_logs_folder, lang='bg')
            if audio_file_path:
                # Inlined logic from play_text_audio
                if os.path.exists(audio_file_path):
                    chat_output.append_stdout(f"üé∂ Now playing: {os.path.basename(audio_file_path)}\n")
                    display(Audio(audio_file_path, autoplay=True))
                else:
                    chat_output.append_stdout(f"‚ùå Audio file not found at '{audio_file_path}' for playback.\n")
            else:
                chat_output.append_stdout("‚ùå Failed to generate audio for Gemini's response.\n")

        except Exception as e:
            chat_output.append_stdout(f"‚ùå Error communicating with Gemini: {e}\n")

# Attach the handler to the button
send_button.on_click(on_send_button_clicked)

# Display the chat interface
chat_interface = widgets.VBox([
    widgets.HTML('<h3>Gemini Chat Interface</h3>'),
    chat_input,
    send_button,
    chat_output
])

display(chat_interface)

print("Gemini Chat interface is ready. Type your questions or prompts and click 'Send to Gemini'.")
print(f"You can now use /save [filename] [content] and /read [filename] to interact with '{project_folder_path}'")
print(f"You can also call 'play_show()' to generate an episode. Gemini's responses will be saved to '{chat_logs_folder}' and read aloud.")
````
